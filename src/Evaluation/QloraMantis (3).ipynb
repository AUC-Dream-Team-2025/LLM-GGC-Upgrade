```python
import matplotlib.pyplot as plt
from transformers import Pix2StructProcessor, Pix2StructForConditionalGeneration
from PIL import Image
import json

# Initialize variables
previous_file_name = ""
results = []

# Initialize loss tracking
losses = []
epochs = []

# Counters for accuracy
correct_human = 0
total_human = 0
correct_augmented = 0
total_augmented = 0

# Define input files for training and testing
train_files = {
    "human": "train_human.json",
    "augmented": "train_augmented.json"
}

test_files = {
    "human": "test_human.json",
    "augmented": "test_augmented.json"
}

# Training phase (1 epoch)
print("Starting training phase...")
for epoch in range(1, 2):  # Train for 1 epoch
    epoch_loss = 0  # Track total loss for the epoch
    epoch_correct = 0  # Track correct predictions for the epoch
    epoch_total = 0  # Track total predictions for the epoch

    # Iterate through both training datasets
    for data_type, input_file in train_files.items():
        print(f"Processing {data_type} training data from {input_file}...")
        with open(input_file, "r") as f:
            data = json.load(f)

        # Iterate through each entry in the dataset
        for entry in data:
            file_name = entry["imgname"]
            file_path = f"train/{file_name}"  # Replace with the actual path to your training images
            question = entry["query"]
            expected_answer = entry["label"]

            # Reload pipeline if the file name changes
            if file_name != previous_file_name:
                print(f"Loading new image: {file_name}")
                previous_file_name = file_name  # Update tracking variable

            # Open the image and verify it's valid
            try:
                image = Image.open(file_path)
                width, height = image.size  # Validate image dimensions (width and height must not be None)
                if width is None or height is None:
                    print(f"Invalid image dimensions for {file_path}")
                    continue  # Skip this image if dimensions are invalid
            except (IOError, SyntaxError) as e:
                print(f"Error loading image {file_path}: {e}")
                continue  # Skip this image and move to the next one

            # Process the image and query
            inputs = processor(images=image, text="Using the chart provided, answer the question: {}".format(question), return_tensors="pt")
            predictions = model.generate(**inputs, max_new_tokens=512)
            generated_answer = processor.decode(predictions[0], skip_special_tokens=True)

            print("Question:", question)
            print("Generated Answer:", generated_answer)
            print("Expected Answer:", expected_answer)

            # Calculate accuracy
            is_correct = generated_answer.strip().lower() == expected_answer.strip().lower()
            epoch_total += 1
            if is_correct:
                epoch_correct += 1

            # Simulate a loss metric (e.g., string similarity or token overlap)
            loss = 1.0 if not is_correct else 0.0  # Loss is 1 if incorrect, 0 if correct
            epoch_loss += loss

            # Track results for training
            results.append({
                "data_type": data_type,
                "imgname": file_name,
                "query": question,
                "label": expected_answer,
                "generated_answer": generated_answer,
                "is_correct": is_correct,
                "loss": loss
            })

    # Track epoch loss and accuracy
    if epoch_total > 0:  # Ensure no division by zero
        losses.append(epoch_loss / epoch_total)  # Average loss for the epoch
        print(f"Epoch {epoch}: Loss = {losses[-1]:.4f}, Accuracy = {epoch_correct / epoch_total * 100:.2f}%")
    else:
        losses.append(0)  # Default loss value if no predictions were made
        print(f"Epoch {epoch}: No predictions made, Loss = 0, Accuracy = 0.00%")
    epochs.append(epoch)

# Testing phase
print("\nStarting testing phase...")
test_loss = 0
test_correct = 0
test_total = 0

for data_type, input_file in test_files.items():
    print(f"Processing {data_type} test data from {input_file}...")
    with open(input_file, "r") as f:
        test_data = json.load(f)

    for entry in test_data:
        file_name = entry["imgname"]
        file_path = f"test/{file_name}"  # Replace with the actual path to your testing images
        question = entry["query"]
        expected_answer = entry["label"]

        # Open the image and verify it's valid
        try:
            image = Image.open(file_path)
        except (IOError, SyntaxError) as e:
            print(f"Error loading image {file_path}: {e}")
            continue  # Skip this image and move to the next one

        # Process the image and query
        inputs = processor(images=image, text="Using the chart provided, answer the question: {}".format(question), return_tensors="pt")
        predictions = model.generate(**inputs, max_new_tokens=512)
        generated_answer = processor.decode(predictions[0], skip_special_tokens=True)

        print("Question:", question)
        print("Generated Answer:", generated_answer)
        print("Expected Answer:", expected_answer)

        # Calculate accuracy
        is_correct = generated_answer.strip().lower() == expected_answer.strip().lower()
        test_total += 1
        if is_correct:
            test_correct += 1

        # Simulate a loss metric
        loss = 1.0 if not is_correct else 0.0
        test_loss += loss

# Calculate overall test accuracy
test_accuracy = test_correct / test_total if test_total > 0 else 0
print(f"Test Loss: {test_loss / test_total if test_total > 0 else 0:.4f}")
print(f"Test Accuracy: {test_accuracy * 100:.2f}%")

# Save results to a JSON file
output_file = "output_results.json"
with open(output_file, "w") as f:
    json.dump(results, f, indent=4)

print(f"Results saved to {output_file}")

# # Plot losses over epochs
# plt.figure(figsize=(10, 6))
# plt.plot(epochs, losses, label="Loss", marker='o')
# plt.xlabel("Epochs")
# plt.ylabel("Loss")
# plt.title("Loss Over Time")
# plt.legend()
# plt.grid()
# plt.show()
```
